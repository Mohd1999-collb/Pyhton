```python
import matplotlib.pyplot as plt


'''plt.subplot(nrows, ncols, index):
nrows: Number of rows in the grid.
ncols: Number of columns in the grid.
index: Position of the current subplot (starting from 1 for the top left).
For example, plt.subplot(2, 2, 1) would create a 2x2 grid and select the first cell for plotting.'''

# Create a 2x2 grid of subplots
plt.figure(figsize=(15, 8), facecolor="brown")

# First subplot
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9], color="red")
plt.title("Line Plot")

# Second subplot
plt.subplot(2, 2, 2)
plt.bar(['A', 'B', 'C'], [3, 7, 5], color='skyblue')
plt.title("Bar Plot")

# Third subplot
plt.subplot(2, 2, 3)
plt.scatter([1, 2, 3], [9, 6, 3], color='orange')
plt.title("Scatter Plot")

# Fourth subplot
plt.subplot(2, 2, 4)
plt.hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4], bins=5, color='lightgreen')
plt.title("Histogram")

# Adjust layout to prevent overlap
plt.tight_layout()

# Saving the figure
plt.savefig("subplot.png", facecolor="yellow")
# Show the plot
plt.show()
```
