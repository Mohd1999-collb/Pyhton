```python
# Legend in Matplotlib
import pandas as pd

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]
y3 = [1, 8, 27, 64, 120]

plt.figure(figsize=(15,5))

plt.plot(x, y1, label="Line plot between Price and Rating", color="red")
plt.plot(x, y2, label="Line plot between Price and Number Cores", color="green")
plt.plot(x, y3, label="Line plot between Price and Preocessor Speed", color="blue")
plt.title("Line plot of data")
plt.legend(["a1","a2","a3"], ncols=3, bbox_to_anchor=(1, 1.1))
plt.show()
```
