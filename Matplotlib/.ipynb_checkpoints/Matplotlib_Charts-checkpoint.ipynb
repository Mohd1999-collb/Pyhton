```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("C:/Users/lenovo/Desktop/AccioJob/Assesment/Python/Pandas/smartphones.csv")


# 1. Bar Graph

gp = data.groupby(data["brand_name"])["model"].count()
x = gp.index[:35]
y = gp.values[:35]
barcolors = ["AliceBlue", "Coral", "DarkOrange", "ForestGreen", "Gold", "LightSeaGreen", "MediumPurple", "NavajoWhite", "OliveDrab", 
             "PeachPuff", "PowderBlue", "RosyBrown", "SlateBlue", "Tomato", "Violet", "YellowGreen", "AntiqueWhite", "Aquamarine", 
             "BlueViolet", "CadetBlue", "Chartreuse", "Crimson", "DeepSkyBlue", "FloralWhite", "Gainsboro", "HotPink", "Indigo", "Lavender", 
             "MediumSeaGreen", "Orchid", "Salmon", "SandyBrown", "SeaShell", "SpringGreen", "Tan"]

plt.figure(figsize=(15,5))
plt.bar(x, y, color=barcolors, width=0.9, edgecolor='Black', label='Mobile Data')
plt.xticks(rotation=90)
plt.xlabel("Brand Name", color="BlueViolet", fontsize=15, fontweight="bold")
plt.ylabel("Count of Models", color="CadetBlue", fontsize=15, fontweight="bold")
plt.title("Count of Models by Brand Name", color="DeepSkyBlue", fontsize=20, fontweight="bold")
plt.grid(color="khaki")
plt.legend()
# plt.tight_layout()  # Adjusts layout to prevent overlap
plt.savefig('sales_by_category.png', format='png', dpi=300, bbox_inches='tight')
plt.show()

```

```python
# 2. Line Plot

gp = data.groupby('brand_name')["processor_brand"].count()
x = gp.index[:35]
y = gp.values[:35]

plt.figure(figsize=(15, 5))
plt.plot(x, y, marker="*", color="ForestGreen", linestyle=":", label="Mobile Data", linewidth=3)
plt.xlabel("Brand Name")
plt.ylabel("Count of Processor by Brand Name")
plt.title("Count of Processor by Brand Name", color="DeepSkyBlue", fontsize=20, fontweight="bold")
plt.xticks(rotation=90)
plt.grid(linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

```python
# 3. Scatter Graph

import numpy as np
x = data["price"]
y = data["rating"]
plt.figure(figsize = (15,5))
colors = np.random.rand(980)
plt.scatter(y, x, c=colors, marker='*', s=200, alpha=0.5, cmap="viridis")
plt.title('Relationship Between Age and Income', fontsize=16, fontweight='bold')
plt.xlabel("Price", fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.colorbar()
plt.legend(["Mobile Data"])
plt.show()
```

```python
# 4. Pie Chart

plt.figure(figsize = (40,8))
gp = data.groupby("brand_name")["processor_speed"].count()
x = gp[10:15].values
label = gp[10:15].index
ex = [0.1, 0, 0, 0, 0.1] # Only "explode" the last slice
slicecolor = ["DarkOrange", "ForestGreen", "Gold", "LightSeaGreen", "MediumPurple"]
style = {'color': 'Chartreuse', 'fontsize': 15, 'fontweight': 'bold'}
plt.pie(x, labels=label, colors=slicecolor, autopct="%1.2f", shadow=True, startangle=60, explode=ex, textprops=style)
plt.title('Pie Chart of Brand Name', fontsize=20, fontweight="bold", color="Gold")
plt.legend()
plt.show()
```

```python
# 5. Box Plot

plt.figure(figsize = (15,7))
gp = data.groupby("brand_name")["processor_speed"].count()
x = gp[10:55].values
y = gp[10:30].values
plt.boxplot([x,y],label=["10-55", "10-120"])
plt.xlabel("Box Plot", fontsize=20, fontweight="bold", color="Chartreuse")
plt.ylabel("Count of Processor by Brand Name", fontsize=15, fontweight="bold", color="Chartreuse")
plt.title("Processor by Brand Name", fontsize=20, fontweight="bold", color="Chartreuse")
plt.legend(loc="upper left", fontsize=10, title="Data", shadow=True, frameon=True, facecolor="lightyellow", edgecolor="black")
plt.grid()
plt.show()
```

```python
# 6. Histogram Plot

# Generating sample data (e.g., ages of a group of people)
data = np.random.normal(30, 10, 50)  # mean=30, std=10, size=1000
plt.figure(figsize = (15, 5))
plt.hist(data, bins=20, color="Chartreuse", edgecolor="black", cumulative=1, density=True, histtype="barstacked")
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Density Histogram of Ages")
plt.show()
```

```python
# 7. Violin Plot

# Generating sample data (e.g., ages of a group of people)
data = np.random.normal(30, 10, 50)
plt.figure(figsize = (15, 5))
plt.violinplot(data, showmedians=True, showmeans=True, showextrema=True)
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Density Histogram of Ages")
plt.show()
```

```python
# 8. Stem Plot

# Sample data
x = np.linspace(0, 10, 10)  # Generate 10 equally spaced values from 0 to 10
y = np.sin(x)                # Compute the sine of each value

# Create the stem plot
plt.figure(figsize=(10, 6))
markerline, stemlines, baseline = plt.stem(x, y, basefmt=" ")
plt.stem(x, y, label='Data Points', basefmt="C2-", markerfmt="*", linefmt="--")
# Add labels and title
plt.title("Stem Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Show the plot
plt.grid(True)  # Optional: add a grid for better visibility
plt.show()
```

```python
# 9. Stack Plot
# Sample data
x = np.arange(1, 6)  # X-axis values
y1 = np.array([2, 3, 4, 5, 6])  # First component
y2 = np.array([1, 2, 3, 2, 1])  # Second component
y3 = np.array([3, 1, 4, 3, 2])  # Third component

# Stack the data
plt.figure(figsize=(15, 6))
colors = ['#ff9999','#66b3ff','#99ff99'] 
plt.stackplot(x, y1, y2, y3, labels=['Component 1', 'Component 2', 'Component 3'], alpha=1, colors=colors, edgecolor="black")

# Adding labels and title
plt.title('Stacked Area Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='upper left')

# Show the plot
plt.grid()  # Optional: add grid lines for better visibility
plt.show()
```

```python
# 10. Step Plot

# Sample data
x = np.arange(1, 11)  # X-axis values from 1 to 10
y = [1, 2, 1, 3, 2, 4, 3, 5, 4, 5]  # Y-axis values with step changes

# Create the step plot
plt.figure(figsize=(15, 5))
plt.step(x, y, where='post', color='red', linestyle='--', linewidth=5, label='Sample Step Plot')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Step Plot Example')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
