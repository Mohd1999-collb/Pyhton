```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Line Plot

plt.figure(figsize=(15,5), facecolor="green")
data = sns.load_dataset("tips") # Dataset of seaborn 
sns.lineplot(data=data, x="day", y="size", palette="viridis", linewidth=5, hue="sex")
plt.show()
```

```python
# 2. Bar Plot

plt.figure(figsize=(15,5), facecolor="green")
color=sns.color_palette("viridis")
data = sns.load_dataset("tips") # Dataset of seaborn 
sns.barplot(data=data, x="day", y="tip", errorbar=None, hue="sex", palette="magma")
plt.title('Customized Bar Plot', fontsize=16, color='darkblue',)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Tip', fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

```python
# 3. Scatter Plot

plt.figure(figsize=(15,5), facecolor="green")
color=sns.color_palette("viridis")
data = sns.load_dataset("tips") # Dataset of seaborn 
sns.scatterplot(data=data, x="day", y="tip", hue="sex", palette="spring", marker="d", s=200, edgecolor='black')
plt.title('Customized Scatter Plot', fontsize=16, color='darkblue',)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Tip', fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

```python
# 4. Heat Map
import numpy as np


plt.figure(figsize=(15,5), facecolor="green")
# Generate a random dataset
data = np.random.rand(5, 5)
sns.heatmap(data=data, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.8, linecolor="black")
plt.xlabel("Row")
plt.ylabel("Column")
plt.show()
```

```python
# 5. Count Plot

plt.figure(figsize=(15,5), facecolor="blue")
# Example dataset
data = sns.load_dataset("titanic")  # Titanic dataset is built into Seaborn

# Basic count plot
sns.countplot(data=data, x="class", hue="who", palette="pastel" , edgecolor="black")  # 'class' is a categorical variable in Titanic data
plt.xlabel("Class")
plt.ylabel("Count of each class")
plt.show()
```

```python
# 6. Pir Plot

# Load example dataset
data = sns.load_dataset("iris")

# Create a pair plot
sns.pairplot(data=data, hue="species",  palette="viridis", height=2.5, diag_kind="hist", markers=["o", "s", "D"])
plt.show()
```

```python
# 7. Strip Plot

plt.figure(figsize=(15,8), facecolor="khaki")
sns.stripplot(data, x="day", y="total_bill", marker="d", jitter=0.3, dodge=True, hue="smoker", palette="magma", size=10)
plt.xlabel("Day", fontsize=20, color="green")
plt.ylabel("Total Bill", fontsize=20, color="green")
plt.title("Strip Plot Between Day & Total Bill", fontsize=20, color="orange")
plt.show()
```

```python
# 7. Bax Plot

plt.figure(figsize=(15,8), facecolor="Tomato")
sns.boxplot(data, x="total_bill", y="day", orient="horizontal", hue="smoker", palette="spring", width=0.5, linewidth=3.8, showfliers=False)
plt.xlabel("Day", fontsize=20, color="green")
plt.ylabel("Total Bill", fontsize=20, color="green")
plt.title("Strip Plot Between Day & Total Bill", fontsize=20, color="orange")
plt.show()

```

```python
# 7. Cat Plot --> Catrgorical Plot

sns.catplot(data=data, x="day", y="total_bill", kind="box", hue="sex", palette="Set2", col="time", height=5, aspect=1.5)
plt.xlabel("Day", fontsize=20, color="green")
plt.ylabel("Total Bill", fontsize=20, color="green", )
plt.title("Strip Plot Between Day & Total Bill", fontsize=20, color="orange")
plt.show()
```

```python
# 8. Multiple Plot

day_order = ["Thur", "Fri", "Sat", "Sun"]

a = sns.FacetGrid(data=data, col="day")
a.map(sns.barplot, "day", "tip", color="red", order=day_order)

# Setting labels for the axes
a.set_axis_labels("Day", "Tip")
a.set_titles("Day: {col_name}")  # Title for each facet
a.fig.suptitle("Strip Plot Between Day & Total Bill", fontsize=20, color="orange", y=1.05)

plt.show()
```

```python
# 9. Relational Plot

sns.relplot(data=data, x="tip", y="total_bill", hue="sex", kind="line", col="day",  style="time", marker="o")
plt.show()
```

```python
# 10. Swarm Plot

# Create a swarm plot
plt.figure(figsize=(15, 5))
sns.swarmplot(data=data, x="day", y="total_bill", hue="sex", palette="Set2", dodge=True)

# Set plot labels and title
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Swarm Plot of Total Bill by Day, with Gender as Hue")
plt.show()
```

```python
# 11. Kernal Density Estimate(KDE) Plot

# Create a swarm plot
plt.figure(figsize=(15, 5))
sns.kdeplot(data=data, x="total_bill", hue="day", palette="Set2", fill=True, multiple="stack")

# Set plot labels and tit
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.title("KDE Plot of Total Bill")
plt.show()
```
